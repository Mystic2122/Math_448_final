```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
df = pd.read_json("NBA_player_data.json")
```

```python
position_mapping = {"PG": 1, "SG": 2, "SF": 3, "PF": 4, "C": 5}
df["Pos"] = df["Pos"].map(position_mapping)
```

```python
positions = {1:"PG", 2:"SG", 3:"SF", 4:"PF", 5:"C"} 
```

```python
#Points
ppg_by_pos = [df[df['Pos'] == pos]['PTS'] for pos in positions]

plt.boxplot(ppg_by_pos, labels=positions.values())
plt.xlabel('Position')
plt.ylabel('Average PPG')
plt.title('Average Points per Game by Position')

plt.xticks(list(positions.keys()), list(positions.values()))

plt.show()
```

```python
#Assists
ast_by_pos = [df[df['Pos'] == pos]['AST'] for pos in positions]

plt.boxplot(ast_by_pos, labels=positions.values(),medianprops={"color":"green"})
plt.xlabel('Position')
plt.ylabel('Average AST')
plt.title('Average Assists per Game by Position')

plt.xticks(list(positions.keys()), list(positions.values()))

plt.show()
```

```python
#Rebounds
total_rbs = [df[df['Pos'] == pos]['TRB'] for pos in positions]

plt.boxplot(total_rbs, labels=positions.values(),medianprops={"color":"green"})
plt.xlabel('Position')
plt.ylabel('Total Rebounds')
plt.title('Average Total Rebounds per Game by Position')

# Customize the tick labels on the x-axis
plt.xticks(list(positions.keys()), list(positions.values()))

plt.show()
```

```python
#Blocks
bpg_by_pos = [df[df['Pos'] == pos]['BLK'] for pos in positions]

plt.boxplot(bpg_by_pos, labels=positions.values())
plt.xlabel('Position')
plt.ylabel('Average BPG')
plt.title('Average Blocks per Game by Position')

plt.xticks(list(positions.keys()), list(positions.values()))

plt.show()

```

```python
#Steals
spg_by_pos = [df[df['Pos'] == pos]['STL'] for pos in positions]

plt.boxplot(spg_by_pos, labels=positions.values())
plt.xlabel('Position')
plt.ylabel('Average SPG')
plt.title('Average Steals per Game by Position')

plt.xticks(list(positions.keys()), list(positions.values()))

plt.show()
```

```python
#Turnovers
tov_by_pos = [df[df['Pos'] == pos]['TOV'] for pos in positions]

plt.boxplot(tov_by_pos, labels=positions.values())
plt.xlabel('Position')
plt.ylabel('Average TOV')
plt.title('Average Turnovers per Game by Position')

plt.xticks(list(positions.keys()), list(positions.values()))

plt.show()
```
